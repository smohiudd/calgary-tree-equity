```python
import pandas as pd
import geopandas as gpd
import glob
import re
```

```python
# Combine individual tree canopy geojson file sinto a single file with property for each year
```

```python
geojson = glob.glob("./*.json")

dfs = []
for file in geojson:   
    year = re.findall('\d+',file)[0] # get year as string from filename

    df = gpd.read_file(file)
    df["frac"] = df["frac"]*100
    df = df.rename(columns={"frac": year})
    # df = df.set_index(['CTUID','geometry','DGUID','PRUID'])
    df = df.set_index(['name','geometry'])
    dfs.append(df)
 
```

```python
tree_canopy = pd.concat(dfs,axis=1, join='inner').round(2)
```

```python
tree_canopy = tree_canopy.reset_index().set_geometry("geometry")
```

```python
tree_canopy.to_file("Tree_canopy.geojson", driver="GeoJSON")  
```
