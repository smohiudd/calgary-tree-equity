## Lidar Data


```python
def calc_canopy(x):
    
    df = gpd.read_file(
        "Trees_2022/geo_export_dc81e6e2-f4a7-4c85-8a9c-deabecc1015d.shp",
        mask=x.geometry
    )
    
    df = df.to_crs('EPSG:32611')
    tree_area = df.area.sum()
    return tree_area/x.geom_reproj.area
    
census_tracts_clipped["geom_reproj"] = census_tracts_clipped.to_crs('EPSG:32611').geometry
census_tracts_clipped["frac"] = census_tracts_clipped.apply(calc_canopy,axis=1)
```

```python
dst_crs = pyproj.CRS('EPSG:32611')

with rasterio.open("tree_lidar_2022_CT.tif") as src:
    
    transform_, width, height = calculate_default_transform(
        src.crs, 
        dst_crs, 
        src.width, 
        src.height, 
        *src.bounds
    )
    
    dst = np.empty((1,height,width))
    
    reproject(
        source=src.read(1),
        destination=dst,
        src_transform=src.transform,
        src_crs=src.crs,
        dst_transform=transform_,
        dst_crs=dst_crs,
        resampling=Resampling.nearest,
        dst_nodata=0
    )
```

```python
fig, ax = plt.subplots(figsize=(20, 10), dpi=200)
ax.set_axis_off()
cmap_ = colors.LinearSegmentedColormap.from_list("", ["white","#008c46"])

show(dst,ax=ax, transform=transform_, cmap=cmap_)

geom_boundary = transform(project, boundary_data.geometry[0])
x,y = geom_boundary.exterior.xy
ax.plot(x,y, linewidth=0.5, c='green')

plt.savefig('lidar.png', bbox_inches='tight')
plt.close(fig)
```
